```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns  
```

```python
#Import data from csv file and create x_axis sequential number series for x axis.
data_signals = np.genfromtxt ('10_project_data_signals.csv', delimiter=",", dtype=float, skip_header=1)
da = pd.read_csv('10_project_data_annotation.csv')
x_axis = np.arange(start=0, stop=len(data_signals), step=1)
```

```python
#Helper functions
#Fill boxplot box with grey color.
def changeFaceColor(bp):
    for box in bp['boxes']:
        box.set(facecolor = 'grey')
        
def start(index):
    return da.loc[index]['start']

def stop(index):
    return da.loc[index]['stop']

first = np.array([start(0),start(1),start(1),start(1),stop(1),stop(1),stop(0)])
second = np.array([stop(1),start(2),start(2),start(2),stop(2),stop(2),stop(0)])
third = np.array([start(3), start(4), start(4),start(4),stop(4),stop(4),stop(3)])
fourth = np.array([stop(4), start(5), start(5),start(5),stop(5),stop(5),stop(3)])
fifth = np.array([stop(5), start(6), start(6), start(6),stop(6),stop(6),stop(3)])
sixth = np.array([start(7), start(8), start(8),start(8),stop(8),stop(8),stop(7)])
```

### Figure 1.1

```python
fig = plt.figure()
#Split plot to 5 horizontal graph with 0 padding
gs = fig.add_gridspec(5, hspace=0)
#Each of 5 plot gonna use same x axis
axs = gs.subplots(sharex=True)

fig.set_size_inches(18,12)
#Plot first column of signals
axs[0].plot(x_axis, data_signals[:,0],color='#575553', label = "P1")
axs[0].set_ylabel("P1")
#All plots has to start from 0 and finish 20000
axs[0].set_xlim(xmin=0)
axs[0].set_xlim(xmax=20000)

#Plot second column of signals
axs[1].plot(x_axis, data_signals[:,1],  color='#575553', label = "P2")
axs[1].set_ylabel("P2")
#Plot third column of signals
axs[2].plot(x_axis, data_signals[:,2], color='#575553', label = "P3")
axs[2].set_ylabel("P3")
#Plot fourth column of signals
axs[3].plot(x_axis, data_signals[:,3], color='#575553', label = "P4")
axs[3].set_ylabel("P4")

#Custom position as a group for boxplot and hide fliers fill box
bp = axs[4].boxplot(first,0,'',vert=False,patch_artist =True, positions=[0.8])
#Fill box color as grey
changeFaceColor(bp)
bp = axs[4].boxplot(second,0,'',vert=False,patch_artist =True, positions=[0.8])
changeFaceColor(bp)
bp = axs[4].boxplot(third,0,'',vert=False,patch_artist =True, positions=[0.2])
changeFaceColor(bp)
bp = axs[4].boxplot(fourth,0,'',vert=False,patch_artist =True, positions=[0.2])
changeFaceColor(bp)
bp = axs[4].boxplot(fifth,0,'',vert=False,patch_artist =True, positions=[0.2])
changeFaceColor(bp)
bp = axs[4].boxplot(sixth,0,'',vert=False,patch_artist =True, positions=[0.8])
changeFaceColor(bp)
axs[4].set_yticks([0.2, 0.8])
axs[4].set_yticklabels(['-','+'])
axs[4].set_xlabel("Genomic Posistion")
axs[4].set_ylabel('Annotation')
# Hide x labels and tick labels for all but bottom plot.
for ax in axs:
    ax.label_outer()
```

### Figure 1.2

```python
fig = plt.figure()
gs = fig.add_gridspec(2, hspace=0)

axs = gs.subplots()
fig.set_size_inches(20,5)

x_axis_labels = [0,  2500,  5000,  7500, 10000, 12500, 15000, 17500,20000]
heat_data = pd.DataFrame(data_signals, columns=["P1", "P2", "P3", "P4"])
hm = sns.heatmap(data = heat_data.T,vmin=0, vmax=1,cmap="Greys",  ax=axs[0], cbar=False)

bp = axs[1].boxplot(first,0,'',vert=False,patch_artist =True, positions=[0.8])
changeFaceColor(bp)
bp = axs[1].boxplot(second,0,'',vert=False,patch_artist =True, positions=[0.8])
changeFaceColor(bp)
bp = axs[1].boxplot(third,0,'',vert=False,patch_artist =True, positions=[0.2])
changeFaceColor(bp)
bp = axs[1].boxplot(fourth,0,'',vert=False,patch_artist =True, positions=[0.2])
changeFaceColor(bp)
bp = axs[1].boxplot(fifth,0,'',vert=False,patch_artist =True, positions=[0.2])
changeFaceColor(bp)
bp = axs[1].boxplot(sixth,0,'',vert=False,patch_artist =True, positions=[0.8])
changeFaceColor(bp)
axs[1].set_xlabel("Genomic Posistion")
axs[1].set_yticks([0.2, 0.8])
axs[1].set_yticklabels(['-','+'])

axs[1].set_ylabel('Annotation')
plt.xticks(x_axis_labels)
# displaying the plotted heatmap
plt.show()
```

### 2. Visualization exercises

```python
colnames=['bar', 'c1', 'c2', 'control'] 
data_scatter = pd.read_csv('10_project_data_scatter.csv')
data_bar = pd.read_csv('10_project_data_barplot.csv',names=colnames,header=0)

fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(3, 2)

fig.set_size_inches(12,10)

colors = {'Condition':'black', 'Control':'red'}
color_label = list(colors.keys())


#Scatter plot
ax1 = fig.add_subplot(gs[0, 0])
ax1.scatter(data_scatter['x1'], data_scatter['x2'], s=80, facecolors='none', edgecolors='#2e2e2d')
ax1.set_xlabel("X1")
ax1.set_ylabel("X2")

#Bar plot
labels = ['X->Y','X->Z','Y->X', 'Y->Z']
x_range = np.arange(len(labels))

width = 0.2
ax2 = fig.add_subplot(gs[0, -1])
ax2.bar(x_range - width,data_bar['c1'], width,color=colors['Condition'])
ax2.bar(x_range ,data_bar['c2'], width,color=colors['Condition'])
ax2.bar(x_range + width ,data_bar['control'], width,color=colors['Control'])
ax2.set_xticks(x_range,labels)
ax2.set_ylabel("Number of events")
       
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in color_label]
ax2.legend(handles,color_label)


#Take whole 2nd row for heatmap
ax3 = fig.add_subplot(gs[1, :])
heat_data = pd.DataFrame(data_signals, columns=["P1", "P2", "P3", "P4"])
hm = sns.heatmap(data = heat_data.T,vmin=0, vmax=1,cmap="Greys",  ax=ax3, cbar=False)
ax3.label_outer()

#Take whole 3th row as a grid for boxplot
ax4 = fig.add_subplot(gs[2, :])
bp = ax4.boxplot(first,0,'',vert=False,patch_artist =True, positions=[0.8])
changeFaceColor(bp)
bp = ax4.boxplot(second,0,'',vert=False,patch_artist =True, positions=[0.8])
changeFaceColor(bp)
bp = ax4.boxplot(third,0,'',vert=False,patch_artist =True, positions=[0.2])
changeFaceColor(bp)
bp = ax4.boxplot(fourth,0,'',vert=False,patch_artist =True, positions=[0.2])
changeFaceColor(bp)
bp = ax4.boxplot(fifth,0,'',vert=False,patch_artist =True, positions=[0.2])
changeFaceColor(bp)
bp = ax4.boxplot(sixth,0,'',vert=False,patch_artist =True, positions=[0.8])
changeFaceColor(bp)
ax4.set_xlabel("Genomic Posistion")
ax4.set_xlim(xmin=0)
ax4.set_xlim(xmax=20000)
ax4.set_yticks([0.2, 0.8])
ax4.set_yticklabels(['-','+'])
ax4.set_xlabel("Genomic Posistion")
ax4.set_ylabel('Annotation')
```
